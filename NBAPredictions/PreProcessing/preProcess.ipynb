# Pre Process The Database

```python
import pandas as pd
import numpy as np
import random
```

```python
#create df
df = pd.read_csv('nbaDatabase2004-2024.csv')
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
def get_season(date):
    year = date.year
    #NBA season starts in October
    if date.month >= 10:
        return f'{year}-{year+1}'
    else:
        return f'{year-1}-{year}'

#initalize season column
df['season'] = df['date'].apply(get_season)

df
```

```python
#initalize the new columns at appropriate index
df['homeTeam_wins'] = 0
df['homeTeam_losses'] = 0
df['awayTeam_wins'] = 0
df['awayTeam_losses'] = 0
cols = df.columns.tolist()
cols.insert(cols.index('homeTeam_id')+1, cols.pop(cols.index('homeTeam_wins')))
cols.insert(cols.index('homeTeam_wins')+1, cols.pop(cols.index('homeTeam_losses')))
cols.insert(cols.index('awayTeam_id')+1, cols.pop(cols.index('awayTeam_wins')))
cols.insert(cols.index('awayTeam_wins')+1, cols.pop(cols.index('awayTeam_losses')))

df = df[cols]
df
```

```python
print(df.columns)
```

```python
#reverse df
df = df.iloc[::-1].reset_index(drop=True)
df
```

```python
def extract_win_loss(dfwithSeason):    
    #initalize dictionaries to track wins and losses
    win_count = {}
    loss_count = {}
    for index, row in dfwithSeason.iterrows():
        #get homeTeam, awayTeam, winner, season from row
        homeTeam = row['homeTeam']
        awayTeam = row['awayTeam']
        winner = row['winner']
        season = row['season']
        #initalize win and loss counts for teams
        if homeTeam not in win_count:
            win_count[homeTeam] = 0
            loss_count[homeTeam] = 0
        if awayTeam not in win_count:
            win_count[awayTeam] = 0
            loss_count[awayTeam] = 0
        
        #update win/loss counts based on winner
        if winner == homeTeam:
            win_count[homeTeam] +=1
            loss_count[awayTeam] +=1
        elif winner == awayTeam:
            win_count[awayTeam] +=1
            loss_count[homeTeam] +=1

        #assign current win/loss counts
        #subtract 1 from the input as we want their previous to the game
        dfwithSeason.at[index, 'homeTeam_wins'] = max(win_count[homeTeam], 0)
        dfwithSeason.at[index, 'homeTeam_losses'] = max(loss_count[homeTeam], 0)
        dfwithSeason.at[index, 'awayTeam_wins'] = max(win_count[awayTeam], 0)
        dfwithSeason.at[index, 'awayTeam_losses'] = max(loss_count[awayTeam], 0)
    return dfwithSeason    

df = df.groupby('season', group_keys=False).apply(extract_win_loss)
df


```

```python
#insert win% column
#if team has zero win then win percentage is zero
#if team has more than zero wins and zero losses win percentage is 100 - divide by zero error handler
df.insert(df.columns.get_loc('homeTeam_id') + 1, 'homeTeam_win%', np.where(df['homeTeam_wins'] == 0, 0, np.where(df['homeTeam_losses'] == 0, 1.0, df['homeTeam_wins'] / (df['homeTeam_wins'] + df['homeTeam_losses']))))
df.insert(df.columns.get_loc('awayTeam_id') + 1, 'awayTeam_win%', np.where(df['awayTeam_wins'] == 0, 0, np.where(df['awayTeam_losses'] == 0, 1.0, df['awayTeam_wins'] / (df['awayTeam_wins'] + df['awayTeam_losses']))))

columns_of_interest = ['homeTeam_win%', 'homeTeam_wins', 'homeTeam_losses', 'awayTeam_win%', 'awayTeam_wins', 'awayTeam_losses']

first_row = df.iloc[0]
row_index = first_row.name 
print(f"Row {row_index + 1}:") 
for column, value in first_row.items():
    if column in columns_of_interest:
        print(f"  Column: {column}, Value: {value}")
    
#confirm this worked as intended
for i in range(2):
    random_row = random.randint(1, len(df.index))
    print(f"Row {random_row}:")
    row = df.iloc[random_row-1]
    for column, value in row.items():
        if column in columns_of_interest:
            print(f"  Column: {column}, Value: {value}")
    print("-" * 30)
```
